# 6. Removing the outliers

We handled outliers using the IQR method and visualized the data distribution for numerical values. It's important to note that there are no outliers in categorical values. Outliers, which represent rare deviations from typical data points, are shown in the graph outside the box

```python
Q1 = data['calories'].quantile(0.25)
IQR1 = data['calories'].quantile(0.75) + data['calories'].quantile(0.25)
calorie_max_limit = Q1 + 1.5 * IQR1


Q2 = data['protein'].quantile(0.25)
IQR2 = data['protein'].quantile(0.75) + data['protein'].quantile(0.25)
protein_max_limit = Q1 + 1.5 * IQR2



Q3 = data['sugar'].quantile(0.25)
IQR3 = data['sugar'].quantile(0.75) + data['sugar'].quantile(0.25)
sugar_max_limit = Q1 + 1.5 * IQR3


Q4 = data['carbohydrate'].quantile(0.25)
IQR4 = data['carbohydrate'].quantile(0.75) + data['carbohydrate'].quantile(0.25)
carbohydrate_max_limit = Q1 + 1.5 * IQR4

print("Calories Max Limit: ",calorie_max_limit)
print("Protien Max Limit: ",protein_max_limit)
print("Sugar Max Limit: ",sugar_max_limit)
print("Carbohydrate Max Limit: ",carbohydrate_max_limit)
```

```python
data['sugar'] = np.where(
    data['sugar']>sugar_max_limit,
    sugar_max_limit,
    data['sugar']
    )

data['calories'] = np.where(
    data['calories']>calorie_max_limit,
    calorie_max_limit,
    data['calories']
    )

data['carbohydrate'] = np.where(
    data['carbohydrate']>carbohydrate_max_limit,
    carbohydrate_max_limit,
    data['carbohydrate']
    )

data['protein'] = np.where(
    data['protein']>sugar_max_limit,
    sugar_max_limit,
    data['protein']
    )
```

#13. Model development

The problem we have to solve here is a binary prediction problem. We have to predict whether or not recipes have a high traffic. To do this, we can use a logisitic regression model and random forest model.

I will start with the logistic regression but before, we need to choose witch variables to use.

#  Selecting variables

Encoding categorical variables

To use the category variable in a prediction model, it has to be changed into dummy vairables.

```python
data2_dummies=pd.get_dummies(data2['category'], drop_first=True)
print(data2_dummies.head())
```

The response variable that needs to be predicted is the high (boolean) variable that I have created. For the explanatory variables, we start with the variables in data2 but we have to drop recipe that do not contain any relevant information, category that is replaced with the dummies variables, high_traffic which was used to create the response variable.

```python
y=data2['high']
X=pd.concat([data2[['calories', 'carbohydrate', 'sugar', 'protein', 'servings']], data2_dummies], axis=1)
print(X.shape, y.shape)
```

Now that the data is ready, we can split recipes in a train and a test set. The test set will contain 20% of the data and the split is done using the response variable as stratifier to maintain equal proportion of high traffic recipes in both the train and test sets.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=12)
```

As said before, the variability of the variables is from diffent order of magnitude. It is then important to standardize the data before building models. The different variables will then be taken without their units to have similar orders of magnitude

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

#15. Logistic regression

For the first model, I choose logistic regression because it is a model widely used to predict a binary response variable. This model is also easy to use and interpretable.

Creation of the logistic regression model, fitted on the train set. Predition on the test set.

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(X_train_scaled, y_train)
logreg_pred_probas=logreg.predict_proba(X_test_scaled)[:,1]
logreg_pred=logreg.predict(X_test_scaled)
```

#16 Random forest classifier

As a comparison model, I have chosen the random forest classifier as it's generally performs well. Moreover the technique behind the random forest model is based on trees so is different from the technique of logisitic regression. The conclusion could then be differents.

Creation of the random forest model, fitted on the train set. Predition on the test set.

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=13)
rf.fit(X_train_scaled, y_train)
rf_pred_probas=rf.predict_proba(X_test_scaled)[:,1]
rf_pred=rf.predict(X_test_scaled)
```

#17.  Model evaluation

# Accuracy

We can first evaluate both models looking at the accuracy score. This means the percentage of correctly classified recipes. The accuracy of the regression model is higher than that of the random forest model, so the regression model performs better.

For the logistic regression, the accuracy score is : 0.7597

For the random forest, the accuracy score is : 0.7150

```python
from sklearn.metrics import accuracy_score
print('For the logistic regression, the accuracy score is :', accuracy_score(y_test, logreg_pred))
print('For the random forest, the accuracy score is :', accuracy_score(y_test, rf_pred))
```

# Recall and F1 score of models

```python
from sklearn.metrics import recall_score, f1_score

# Calculate recall for Logistic Regression
logreg_recall = recall_score(y_test, logreg_pred)
print('Recall for Logistic Regression:', logreg_recall)

# Calculate F1 score for Logistic Regression
logreg_f1 = f1_score(y_test, logreg_pred)
print('F1 score for Logistic Regression:', logreg_f1)

# Calculate recall for Random Forest
rf_recall = recall_score(y_test, rf_pred)
print('Recall for Random Forest:', rf_recall)

# Calculate F1 score for Random Forest
rf_f1 = f1_score(y_test, rf_pred)
print('F1 score for Random Forest:', rf_f1)
```

# ROC curve

Using the roc curve, the logistic regression model seems to perform better because his curve is furthest from the dashed line.

```python
from sklearn.metrics import roc_curve
logreg_fpr, logreg_tpr, logreg_thresholds=roc_curve(y_test, logreg_pred_probas)
rf_fpr, rf_tpr, rf_thresholds=roc_curve(y_test, rf_pred_probas)
plt.plot(logreg_fpr, logreg_tpr)
plt.plot(rf_fpr, rf_tpr, color='r')
plt.plot([0,1],[0,1], 'k--')
plt.legend(['Logistic regression', 'Random forest'])
plt.show()
```

# ROC AUC

The model with the highest ROC AUC is the logistic regression model. So the logisitic regression model perfoms better than the random forest model on this dataset.

For the logistic regression, the ROC AUC score is : 0.799

For the random forest, the ROC AUC score is : 0.788

```python
from sklearn.metrics import roc_auc_score
print('For the logistic regression, the ROC AUC score is :', roc_auc_score(y_test, logreg_pred_probas))
print('For the random forest, the ROC AUC score is :', roc_auc_score(y_test, rf_pred_probas))
```

# Business metric

As we have to correctly predict high traffic recipes 80% of the time. We need to see if the proportion of TP among the recipies that have a high traffic is at least of 80%. This means that the sensitivity must be higher than 80%

# Evaluation for logistic regression

For the logistic regression model, high traffic recipies are correctly classified 82% of the time

```python
conf_mat=pd.crosstab(y_test, logreg_pred)
print(conf_mat)

TN= conf_mat[0][0]
TP= conf_mat[1][1]
FN= conf_mat[1][0]
FP= conf_mat[0][1]

sensitivity = TP/(FN+TP)
print('The sensitivity score is', sensitivity)
```

# Evaluation for random forest

For the random forest model, high traffic recipes are correctly classified 76% of the time. The logistic regression model is best for this metric too.

```python
conf_mat=pd.crosstab(y_test, rf_pred)
print(conf_mat)

TN= conf_mat[0][0]
TP= conf_mat[1][1]
FN= conf_mat[1][0]
FP= conf_mat[0][1]

sensitivity = TP/(FN+TP)
print('The sensitivity score is', sensitivity)
```
